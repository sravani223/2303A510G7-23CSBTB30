```python
import pandas as pd
df = pd.read_csv('/content/employee_performance.csv')
print("Dataset Head:")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Summary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Monthly_Salary'], bins=10, kde=True, color='blue')
plt.title(" Monthly_Salary")
plt.xlabel(" Years_of_Experience")
plt.ylabel("Performance_Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Years_of_Experience'], y=df['Monthly_Salary'], hue=df['Department'])
plt.title("Monthly Salary vs. Years Of Experience")
plt.xlabel("Years Of Experience")
plt.ylabel("Monthly Salary")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
