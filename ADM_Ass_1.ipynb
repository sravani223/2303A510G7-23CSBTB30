```python
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['Age'] =df['Age'].fillna(df['Age'].mean())
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mean())
df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
```

```python
from sklearn.preprocessing import StandardScaler
s_scaler = StandardScaler()
df[['Work_Hours_Per_Week']] = s_scaler.fit_transform(df[[ 'Work_Hours_Per_Week']])
df['Work_Hours_Per_Week']
```

```python
df['Age'] = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
```

```python
df['Salary'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()
df
```
